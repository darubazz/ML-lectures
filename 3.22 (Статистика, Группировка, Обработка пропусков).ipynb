```python
planets = sns.load_dataset('planets')
planets
```

```python
# Как заполнить пропуски в колонках средним по колонке

for column in planets.columns:
    if planets[column].dtype == 'object':
        continue
        
    col_mean = planets[column].mean()
    planets[column] = planets[column].fillna(col_mean)

pd.isna(planets).sum()
```
