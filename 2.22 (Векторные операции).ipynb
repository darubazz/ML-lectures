# Векторные операции

```python
list_a = [10, 20, 30]
list_b = [1, 2, 3]

arr_a = np.array(list_a)
arr_b = np.array(list_b)

print(list_a + list_b)
print(arr_a ** arr_b)  # Математические операторы с np.ndarray работают ПОЭЛЕМЕНТНО
```

```python
list_a = [10, 20, 30]
list_b = [2, 3, 4]

arr_a = np.array(list_a)
arr_b = np.array(list_b)

print(arr_a - arr_b)
print(arr_a + arr_b)
print(arr_a * arr_b)
print(arr_a / arr_b)
print(arr_a // arr_b)
print(arr_a % arr_b)
print(arr_a ** arr_b)
```

**N.B. Для матричного умножения матриц в numpy используется оператор @**

```python
arr_a = np.array([2, 4, 8, 16])
b = 5

print(arr_a - b)
print(arr_a + b)
print(arr_a * b)
print(arr_a / b)
print(arr_a // b)
print(arr_a % b)
print(arr_a ** b)
```

```python
mat_a = np.array([[2, 4, 8, 16], [1, 2, 3, 4]])
b = 5

print(mat_a - b)
print(mat_a + b)
print(mat_a * b)
print(mat_a / b)
print(mat_a // b)
print(mat_a % b)
print(mat_a ** b)
```

## Broadcasting Rules
1. Если **размерность (количество измерений)** двух операндов различается, то форма (shape) операнда с меньшей размерностью дополняется единицами с левой стороны
1. Если **форма** двух массивов не совпадает в каком-то из измерений и у одного из операндов в этом измерении единица, то этот операнд дублируется вдоль этой оси до совпадения с большим
1. Если форма не совпадает и ни одно из измерений != 1 - то операция запрещена


```python
a = np.random.randint(10, size=5)
b = np.random.randint(10, size=(1, 1, 1, 5))
c = a + b

print(a)
print(b)
print(c)
print(np.broadcast_to(a, b.shape))
```

```python
A = np.zeros((3, 3))

b1 = np.array([1, 2, 3])
b2 = np.array([[1], [2], [3]])

print('A + b1 =\n', A + b1)
print('\nA + b2 =\n', A + b2)
```

# Векторные операции

Заключаются в избегании использования циклов там, где это можно

```python
v1 = np.random.rand(2048 * 2)
v2 = np.random.rand(2048 * 2)

l1 = list(v1)
l2 = list(v2)
```

```python
%%timeit
dotprod = 0
for i in range(len(l1)):
    dotprod += l1[i] * l2[i]
```

```python
%%timeit
dotprod = 0
for x, y in zip(l1, l2):
    dotprod += x * y
```
